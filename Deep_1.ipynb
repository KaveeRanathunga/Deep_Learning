```python
import pandas as pd
dataset = pd.read_csv(r'D:\KAVEESHA\DataScience\DeepLearning\iris.csv').values
print(dataset.shape)
print(dataset)
```

```python
data = dataset[:,0:4]
print(data)
```

```python
target = dataset[:,4]
print(target)
```

# FFNN Architecture


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential() # empty NN

model.add(Input(shape=(4,)))  # Define the input layer shape
model.add(Dense(16, activation='relu'))  # Hidden layer 1
model.add(Dense(40, activation='relu'))  # Hidden layer 2
model.add(Dense(10, activation='relu'))  # Hidden layer 3
model.add(Dense(3, activation='softmax'))  # Output layer (3 classes)

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam',metrics = ['accuracy'])

model.summary()
```

#### we didn't get input layer as a active layer. So we firstly add the first hidden layer.

## Now we are going to seperate data set to train and test data

```python
from tensorflow.keras.utils import to_categorical

new_target = to_categorical(target)
print(new_target)  # show first 5 rows
```

```python
print(new_target.shape)
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_target, test_target = train_test_split(
    data,new_target, test_size=0.2
)
```

```python
print(train_data.shape)
print(test_data.shape)
```

```python
print(test_data)
```

```python
print(test_target)
```

```python
model.fit(train_data,train_target,epochs = 200, validation_data = (test_data,test_target), batch_size = 4, steps_per_epoch = 50)  #training
```

```python
from matplotlib import pyplot as plt

plt.plot(model.history.history['loss'], label='Training Loss')
plt.plot(model.history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(model.history.history['accuracy'], label='Training Accuracy')
plt.plot(model.history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()
plt.show()
```
